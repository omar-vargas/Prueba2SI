```python
import pandas as pd
import matplotlib.pyplot as plt


# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('Situaci_n_V_ctimas_Minas_Antipersonal_en_Colombia_20240103.csv')

# Mostrar las primeras filas del DataFrame
print(df.head())

# Obtener información general sobre el DataFrame
print(df.info())

# Estadísticas descriptivas de los datos
print(df.describe())
```

```python
resumen_categorico = df['estado'].value_counts()
print(resumen_categorico)
resumen_categorico = df['Actividad'].value_counts()
print(resumen_categorico)
```

```python
import pandas as pd

# Suponiendo que tienes un DataFrame llamado df
# ...

# Obtener los nombres de las columnas
nombres_columnas = df.columns.tolist()

# Mostrar los nombres de las columnas
print(nombres_columnas)

```

```python
df.columns = df.columns.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
# Cambiar los nombres de las columnas
df = df.rename(columns={'genero': 'Temp', 'estado': 'genero', 'Temp': 'estado'})




# Análisis de los datos
total_victimas_por_anio = df['ano'].value_counts()
total_victimas_por_mes = df['mes'].value_counts()
total_victimas_por_departamento = df['departamento'].value_counts()
total_victimas_por_actividad = df['Actividad'].value_counts()
total_victimas_por_genero = df['estado'].value_counts()
total_victimas_por_estado = df['sitio'].value_counts()

# Mostrar información resumida en consola
print("Total de víctimas por año:")
print(total_victimas_por_anio)
print("\nTotal de víctimas por mes:")
print(total_victimas_por_mes)
print("\nTotal de víctimas por departamento:")
print(total_victimas_por_departamento)
print("\nTotal de víctimas por actividad:")
print(total_victimas_por_actividad)
print("\nTotal de víctimas por género:")
print(total_victimas_por_genero)
print("\nTotal de víctimas por estado:")
print(total_victimas_por_estado)

# Escribir los resultados en un archivo de texto
with open('resultados.txt', 'w') as file:
    file.write("Total de víctimas por año:\n")
    file.write(total_victimas_por_anio.to_string() + '\n\n')
    file.write("Total de víctimas por mes:\n")
    file.write(total_victimas_por_mes.to_string() + '\n\n')
    file.write("Total de víctimas por departamento:\n")
    file.write(total_victimas_por_departamento.to_string() + '\n\n')
    file.write("Total de víctimas por actividad:\n")
    file.write(total_victimas_por_actividad.to_string() + '\n\n')
    file.write("Total de víctimas por género:\n")
    file.write(total_victimas_por_genero.to_string() + '\n\n')
    file.write("Total de víctimas por estado:\n")
    file.write(total_victimas_por_estado.to_string() + '\n\n')

print("Proceso completado. Los resultados se han guardado en 'resultados.txt'")


```
